```python
import pandas as pd
```

<h2>MFT scores</h2>

```python
models = ['llama3_1_405b_instruct_fp8','deepseek_v3_0324','chatgpt_4o_latest','claude_3_7_sonnet_20250219']

df = pd.read_csv('C:/Users/user/Downloads/Probe Experiments/Probe Experiments/experiment_1_results_mft.csv')

for i in models:
    df[f"{i}_score"] = df[f"{i}_answer"] == df["correct_answer"]
    accuracy = df[f"{i}_score"].sum() / len(df)
    print(f"{i} accuracy: {accuracy:.2%}")
```
